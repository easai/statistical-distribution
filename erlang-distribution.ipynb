```python
import matplotlib.pyplot as plt
import math
import numpy as np
from scipy import stats

x=np.arange(0,4,.01)
l=[]

def erlang(x,k,mean,var):
    y=var**k*x**(k-1)*math.exp(-var*x)/math.factorial(k-1)
    return y

for i in np.linspace(1,5,5):
    #y=[stats.erlang.pdf(x=m, a=i, loc=0, scale=1) for m in x]
    y=[erlang(m,i,0,1) for m in x]
    plt.plot(x,y)
    l.append("n = "+str(i))
plt.legend(l)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title('title')
plt.title("Erlang Distribution PDF")
plt.show()
```

```python
import matplotlib.pyplot as plt
import math
import numpy as np
from scipy import stats

x=np.arange(0,10,.01)
l=[]


for i in np.linspace(1,5,5):
    y=[stats.erlang.cdf(x=m, a=i, loc=0, scale=1) for m in x]
    plt.plot(x,y)
    l.append("n = "+str(i))

plt.legend(l)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title('title')
plt.title("Erlang Distribution CDF")
plt.show()
```
