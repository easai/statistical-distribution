```python
import matplotlib.pyplot as plt
import math
import numpy as np
from scipy import stats

x=np.arange(-5,5,.1)
l=[]

def cauchy(x,x0,gamma):
    return 1/(math.pi*gamma*(1+((x-x0)/gamma)**2))

for i in np.linspace(.5,3,5):
    y=stats.cauchy.pdf(x=x, loc=0, scale=i)
#     y=[cauchy(m, 0, i) for m in x]
    plt.plot(x,y)
    l.append("γ = "+str(i))

plt.legend(l)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title('title')
plt.title("Cauchy Distribution PDF")
plt.show()
```

```python
import matplotlib.pyplot as plt
import math
import numpy as np
from scipy import stats

x=np.arange(-5,5,.1)
l=[]

for i in np.linspace(.5,3,5):
    y=stats.cauchy.cdf(x=x, loc=0, scale=i)
    plt.plot(x,y)
    l.append("γ = "+str(i))

plt.legend(l)
plt.xlabel("x")
plt.ylabel("F(x)")
plt.title('title')
plt.title("Cauchy Distribution CDF")
plt.show()
```
