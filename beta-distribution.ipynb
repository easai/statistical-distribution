```python
import matplotlib.pyplot as plt
import math
import numpy as np
from scipy import stats

x=np.arange(0,1,.01)
l=[]

for i in np.linspace(1.5,3,5):
    y=stats.beta.pdf(x=x, a=i, b=i)
    plt.plot(x,y)
    l.append("a = "+str(i))

plt.legend(l)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title('title')
plt.title("Beta Distribution PDF a=b, 1<a")
plt.show()
```

```python
import matplotlib.pyplot as plt
import math
import numpy as np
from scipy import stats

x=np.arange(0,1,.01)
l=[]

for i in np.linspace(0.1,.9,5):
    y=stats.beta.pdf(x=x, a=1, b=i)
    plt.plot(x,y)
    l.append(f"a=1, b={i:.1f}")

plt.legend(l)
plt.xlabel("x")
plt.ylabel("F(x)")
plt.title('title')
plt.title("Beta Distribution PDF a=1, b<1")
plt.show()
```
